```python
import numpy as np
import sys
import matplotlib.pyplot as plt

sys.path.append(r"C:\Users\haoyuan\Documents\GitHub\CrystalDiff")

from CrystalDiff import util, crystal
```

```python
miscut = np.deg2rad(5.)

A = np.array([2.5, 2.644])
n = np.array([ -np.sin(miscut), -np.cos(miscut)])

# Lattice distance for silicon 220 um
lattice = 192015.569 * 1e-9 
```

```python
energy_list = [9, 9.5, 10, 10.5, 11, 11.5, 12]

C_holder = np.zeros((len(energy_list), 50, 2))
Bx_holder = np.zeros((len(energy_list), 50))
gap_holder = np.zeros((len(energy_list), 50))
d_vals = np.linspace(start=0.5, stop=4.5, num=50)
    
for e_idx in range(len(energy_list)):
    
    # Get bragg angle
    energy = energy_list[e_idx]
    wave_vec_in = util.kev_to_wave_number(energy=energy)

    theta_bragg = np.arcsin( np.pi / lattice / wave_vec_in)

    print("The bragg angle is {:.2f} degree.".format(np.rad2deg(theta_bragg)))

    # Get k vector
    k = np.array([np.cos(theta_bragg), np.sin(theta_bragg)])

    # Loop through all d_vals
    for idx in range(50):

        d = d_vals[idx]

        # Get D
        D = np.array([d, 1.])

        # Get l
        l = np.dot(A - D, n) / np.dot(k, n)

        # Get point C
        C_holder[e_idx, idx] = D + k * l

        # Get the x coordiante of B
        Bx_holder[e_idx, idx] = d + 2 * l * np.cos(theta_bragg)

        # Get the gap
        gap_holder[e_idx, idx] = l * np.sin(theta_bragg)
```

# Gap size

```python
plt.figure(figsize=(10, 5))

for x in range(len(energy_list)):
    plt.plot(d_vals, 
             gap_holder[x,:], 
             label='{}, {:.2f}, {:.2f}'.format(energy_list[x], 
                                       np.max(gap_holder[x]),
                                       np.min(gap_holder[x])))

    print("Energy:{} keV, Gap Max {:.2f} cm, Gap Min {:.2f} cm".format(energy_list[x], 
                                                               np.max(gap_holder[x]),
                                                               np.min(gap_holder[x])))
#plt.ylim([0, 5])
plt.legend()
plt.show()
```

# Calculate the corresponding delay time range

```python
grating_list = [crystal.RectangleGrating(), crystal.RectangleGrating()]
grating_list[0].set_a(0.5)
grating_list[0].set_b(0.5)
grating_list[1].set_a(0.5)
grating_list[1].set_b(0.5)
```

# Start with a specific configuration for10keV

```python
# incident enenrgy keV
energy = 10
wave_vec_in = util.kev_to_wave_number(energy)

# Distance between gratings
d = 9e6

# Angular deviation induced by the grating
delta = np.arctan(grating_list[0].base_wave_vector[1] / wave_vec_in)

# Bragg angle
theta_bragg = np.arcsin( np.pi / lattice / wave_vec_in)

# Get the ideal gap difference
delta_g_ideal = d * np.sin(delta) / np.cos(theta_bragg) / 2.
```

```python
def random():
    return np.random.rand() * 10

g1 = 2.5 * 1e4 # 2.5 cm
g2 = g1 + delta_g_ideal + random() # The ideal g2 with an error of 10 um

# Assume that human are able to deterministically move 
# the four CC to the ideal position with an error of 10 um
f1 = 1.2 * 1e4 + random()
f2 = 1.2 * 1e4 + random()
f3 = 1.3 * 1e4 + random()

f4_ideal = f3 + f2 - f1 - (g2 - g1) + d * np.sin(delta) / np.cos(theta_bragg)

f4 = f4_ideal + random()


print("g1 {:.2f}, g2 {:.2f}".format(g1 / 1e4, g2 / 1e4))
print("f1 {:.2f}, f2 {:.2f}, f3 {:.2f}, f4 {:.2f}".format(f1/1e4, f2/1e4, f3/1e4, f4/1e4))
```

# Optimize the position on the second grating

```python
# Get the position on the second grating from the fixed branch
h_fixed = 2 * (g2 - g1) * np.cos(theta_bragg) / np.cos(delta) - d * np.tan(delta)
h_var = (f2 - f1 + f3 - f4) * np.cos(theta_bragg) * 2 / np.cos(delta) + d * np.tan(delta)
print(h_fixed, h_var)
print("f1 {:.5f}, f2 {:.5f}, f3 {:.5f}, f4 {:.5f}".format(f1/1e4, f2/1e4, f3/1e4, f4/1e4))

# Get the corresponding change in the value of f3 and f4
change = (h_fixed - h_var) / 2 * np.cos(delta) / np.cos(theta_bragg)

# Move the last two channel-cuts
f3 += change / 2.
f4 -= change / 2.

# Calculate the h value for the variable branch
h_var = (f2 - f1 + f3 - f4) * np.cos(theta_bragg) * 2 / np.cos(delta) + d * np.tan(delta)

print(h_fixed, h_var)
print("f1 {:.5f}, f2 {:.5f}, f3 {:.5f}, f4 {:.5f}".format(f1/1e4, f2/1e4, f3/1e4, f4/1e4))
```

# Optimize the path length

```python
path_fixed = (d / np.cos(delta) - 2 * (g2 - g1) * np.cos(theta_bragg) * np.tan(delta) +
              (g1 + g2) / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg)))

path_var = (d * np.cos(delta) + h_var * np.sin(delta) + 
            (f1 + f2 + f3 + f4) / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg)))

print(path_fixed, path_var, (path_fixed - path_var)/util.c)
print("f1 {:.5f}, f2 {:.5f}, f3 {:.5f}, f4 {:.5f}".format(f1/1e4, f2/1e4, f3/1e4, f4/1e4))
print("fmax {:.5f}, fmin {:.5f}".format(np.max(gap_holder[2]),  np.min(gap_holder[2])))
print()

path_difference = path_fixed - path_var

f3 += path_difference * np.sin(theta_bragg) / (1 - np.cos(2 * theta_bragg)) / 2
f4 += path_difference * np.sin(theta_bragg) / (1 - np.cos(2 * theta_bragg)) / 2

path_var = (d * np.cos(delta) + h_var * np.sin(delta) + 
            (f1 + f2 + f3 + f4) / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg)))
print(path_fixed, path_var, (path_fixed - path_var)/util.c)
print("f1 {:.5f}, f2 {:.5f}, f3 {:.5f}, f4 {:.5f}".format(f1/1e4, f2/1e4, f3/1e4, f4/1e4))
print("fmax {:.5f}, fmin {:.5f}".format(np.max(gap_holder[2]),  np.min(gap_holder[2])))
```

# Calculate the delay range this time

```python
d_gap_min = max(f1 - np.min(gap_holder[2]) * 1e4, f2 - np.min(gap_holder[2] * 1e4))
d_gap_max = min(np.max(gap_holder[2])* 1e4 - f1, np.max(gap_holder[2])* 1e4 - 2)

path_var_min = 2 * d_gap_min / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg))
path_var_max = 2 * d_gap_max / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg))

print(path_var_max / util.c / 1000, path_var_min / util.c / 1000)
```

# Do the calculation for different energy

```python
energy_list = [9, 9.5, 10, 10.5, 11, 11.5, 12]

    
for e_idx in range(len(energy_list)):
    
    # Get bragg angle
    energy = energy_list[e_idx]
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    print("Do the calculation for enenrgy {} keV".format(energy))

    wave_vec_in = util.kev_to_wave_number(energy=energy)
    theta_bragg = np.arcsin( np.pi / lattice / wave_vec_in)
    print("The bragg angle is {:.2f} degree.".format(np.rad2deg(theta_bragg)))

    # Get angles
    delta = np.arctan(grating_list[0].base_wave_vector[1] / wave_vec_in)

    ##################################################################
    #     Step 1: Get the vertical position on the fixed branch on the second grating
    ##################################################################
    # Get the position on the second grating from the fixed branch
    h_fixed = 2 * (g2 - g1) * np.cos(theta_bragg) / np.cos(delta) - d * np.tan(delta)
    h_var = (f2 - f1 + f3 - f4) * np.cos(theta_bragg) * 2 / np.cos(delta) + d * np.tan(delta)

    # Get the corresponding change in the value of f3 and f4
    change = (h_fixed - h_var) / 2 * np.cos(delta) / np.cos(theta_bragg)

    # Move the last two channel-cuts
    f3_new = f3 + change / 2.
    f4_new = f4 - change / 2.

    # Calculate the h value for the variable branch
    h_var = (f2 - f1 + f3_new - f4_new) * np.cos(theta_bragg) * 2 / np.cos(delta) + d * np.tan(delta)

    print("vertical positions are:", h_fixed, h_var)
    print("f1 {:.5f}, f2 {:.5f}, f3 {:.5f}, f4 {:.5f}".format(f1/1e4, f2/1e4, f3_new/1e4, f4_new/1e4))
    print("fmax {:.5f}, fmin {:.5f}".format(np.max(gap_holder[e_idx]),  np.min(gap_holder[e_idx])))
    
    ##################################################################
    #     Step 2: Compensate the path length
    ##################################################################
    # Get the path length
    path_fixed = (d / np.cos(delta) - 2 * (g2 - g1) * np.cos(theta_bragg) * np.tan(delta) +
              (g1 + g2) / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg)))

    path_var = (d * np.cos(delta) + h_var * np.sin(delta) + 
                (f1 + f2 + f3_new + f4_new) / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg)))
    print("Relative path length before additional adjustment:", (path_fixed - path_var)/util.c)

    # Get the path difference
    path_difference = path_fixed - path_var

    # Get the new gaps
    f3_new = f3_new + path_difference * np.sin(theta_bragg) / (1 - np.cos(2 * theta_bragg)) / 2
    f4_new = f4_new + path_difference * np.sin(theta_bragg) / (1 - np.cos(2 * theta_bragg)) / 2

    path_var = (d * np.cos(delta) + h_var * np.sin(delta) + 
                (f1 + f2 + f3_new + f4_new) / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg)))
    print("Relative Path lengths:", (path_fixed - path_var)/util.c)
    print("f1 {:.5f}, f2 {:.5f}, f3 {:.5f}, f4 {:.5f}".format(f1/1e4, f2/1e4, f3_new/1e4, f4_new/1e4))
    print("fmax {:.5f}, fmin {:.5f}".format(np.max(gap_holder[e_idx]),  np.min(gap_holder[e_idx])))
    
    ##################################################################
    #     Step 3: Calculate the delay range
    ##################################################################
    d_gap_min = max(f1 - np.min(gap_holder[e_idx]) * 1e4, f2 - np.min(gap_holder[e_idx] * 1e4))
    d_gap_max = min(np.max(gap_holder[e_idx])* 1e4 - f1, np.max(gap_holder[e_idx])* 1e4 - 2)

    path_var_min = 2 * d_gap_min / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg))
    path_var_max = 2 * d_gap_max / np.sin(theta_bragg) * (1 - np.cos(2 * theta_bragg))
    
    print("Delay max: {:.4f} ps".format(path_var_max / util.c / 1000))
    print("Delay min: {:.4f} ps".format(-path_var_min / util.c / 1000))
    
```
