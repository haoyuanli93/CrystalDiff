```python
import numpy as np
import sys
import matplotlib.pyplot as plt

sys.path.append(r"C:\Users\haoyuan\Documents\GitHub\CrystalDiff")

from CrystalDiff import util, crystal
```

```python
miscut = np.deg2rad(5.)

A = np.array([2.5, 2.644])
n = np.array([ -np.sin(miscut), -np.cos(miscut)])

# Lattice distance for silicon 220 um
lattice = 192015.569 * 1e-9 
```

```python
energy_list = [9, 9.5, 10, 10.5, 11, 11.5, 12]

C_holder = np.zeros((len(energy_list), 50, 2))
Bx_holder = np.zeros((len(energy_list), 50))
gap_holder = np.zeros((len(energy_list), 50))
d_vals = np.linspace(start=0.5, stop=4.5, num=50)
    
for e_idx in range(len(energy_list)):
    
    # Get bragg angle
    energy = energy_list[e_idx]
    wave_vec_in = util.kev_to_wave_number(energy=energy)

    theta_bragg = np.arcsin( np.pi / lattice / wave_vec_in)

    print("The bragg angle is {:.2f} degree.".format(np.rad2deg(theta_bragg)))

    # Get k vector
    k = np.array([np.cos(theta_bragg), np.sin(theta_bragg)])

    # Loop through all d_vals
    for idx in range(50):

        d = d_vals[idx]

        # Get D
        D = np.array([d, 1.])

        # Get l
        l = np.dot(A - D, n) / np.dot(k, n)

        # Get point C
        C_holder[e_idx, idx] = D + k * l

        # Get the x coordiante of B
        Bx_holder[e_idx, idx] = d + 2 * l * np.cos(theta_bragg)

        # Get the gap
        gap_holder[e_idx, idx] = l * np.sin(theta_bragg)
```
