```python
import numpy as np
import sys
import time
import matplotlib.pyplot as plt
import matplotlib as mpl

sys.path.append(r"C:\Users\haoyuan\Documents\GitHub\CrystalDiff")

from CrystalDiff import util
```

```python
mpl.rc('font',family='Times New Roman')
```

# Define variables for Diamond

```python
# Speed of light * reduced Planck constant
c_h_bar = 2.99792458 * 6.582119569 * 1e-8

# Define photon energy and wave vector
photon_energy = 10000  #(eV)
k = photon_energy / c_h_bar

# Define the groove width
b = 1e-6  # (m)

# Define the tooth width
a = 1e-6  # (m)

# Define the refraction coefficent
n = 1 - 0.73031 * 1e-5 + 1.j * 0.61521 * 1e-8

# Define a series of height
h_for_diamond = np.arange(1200) * 1e-8   # The unit is m. The smallest tooth height is 10nm.
```

```python
diamond_0= util.get_square_grating_transmission(k = k, m = 0, n = n, h = h_for_diamond, a = a, b = b)
diamond_1= util.get_square_grating_transmission(k = k, m = 1, n = n, h = h_for_diamond, a = a, b = b)
```

```python
plt.figure(figsize=(5, 5))

# Change the x axis unit into um
plt.plot(h_for_diamond * 1e6, diamond_0, 'r')
plt.plot(h_for_diamond * 1e6, diamond_1, 'b')

plt.xlabel("Hight of the tooth in um", fontsize=20)
plt.ylabel("Transmission efficiency", fontsize=20)
plt.legend(["0th order", "1st order"], fontsize=20)


plt.show()
```

# Define variables for Silicon

```python
# Speed of light * reduced Planck constant
c_h_bar = 2.99792458 * 6.582119569 * 1e-8

# Define photon energy and wave vector
photon_energy = 10000  #(eV)
k = photon_energy / c_h_bar

# Define the groove width
b = 1e-6  # (m)

# Define the tooth width
a = 1e-6  # (m)

# Define the refraction coefficent
n = 1 - 0.48816E-05 + 1.j * 0.74355E-07

# Define a series of height
h_for_silicon = np.arange(1800) * 1e-8   # The unit is m. The smallest tooth height is 10nm.
```

```python
silicon_0= util.get_square_grating_transmission(k = k, m = 0, n = n, h = h_for_silicon, a = a, b = b)
silicon_1= util.get_square_grating_transmission(k = k, m = 1, n = n, h = h_for_silicon, a = a, b = b)
```

```python
plt.figure(figsize=(5, 5))

# Change the x axis unit into um
plt.plot(h_for_silicon * 1e6, silicon_0, 'r')
plt.plot(h_for_silicon * 1e6, silicon_1, 'b')

plt.xlabel("Hight of the tooth in um", fontsize=20)
plt.ylabel("Transmission efficiency", fontsize=20)
plt.legend(["0th order", "1st order"], fontsize=20)


plt.show()
```

# Put two image together

```python
plt.figure(figsize=(10, 5))

# Change the x axis unit into um
plt.plot(h_for_diamond * 1e6, diamond_0, 'r')
plt.plot(h_for_diamond * 1e6, diamond_1, 'b')
plt.plot([8.49, 8.49], [0, 1], 'k--')


plt.plot(h_for_silicon * 1e6, silicon_0, 'g--')
plt.plot(h_for_silicon * 1e6, silicon_1, 'y--')
plt.plot([12.7, 12.7], [0, 1], 'k--')

plt.xlabel("Depth of the Grating Groove (um)", fontsize=16)
plt.ylabel("Transmission efficiency", fontsize=16)
plt.legend(["Diamond, 0th",
            "Diamond, 1st",
            'Silicon, 0th',
            'Silicon, 1th'], fontsize=14)

plt.xlim([0, 18])
plt.ylim([0,1.1])

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.title("Transmission Efficiency of Square Grating With Silicon and Diamond", fontsize=14)

plt.show()
```
