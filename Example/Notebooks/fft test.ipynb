```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(-100, 100, 1e5)
b = np.exp(-np.square(a))
```

```python
plt.plot(b)
plt.show()
```

```python
c = np.fft.fftshift(a)

plt.plot(c)
plt.show()
```

```python
num = 10 ** 4
half_num = num // 2
d= np.zeros(num)

d[:half_num] = c[-half_num:]
d[half_num:] = c[:half_num]
```

```python
plt.plot(d)
plt.show()
```

```python
e = np.fft.fftshift(b)
f = np.exp(-np.square(c))
print(np.sum(np.abs(e -f)))
```

```python
spec = np.fft.fft(e)
```

```python
field = np.fft.ifft(spec)
```

```python
assemble  = np.zeros(num, dtype = np.complex128)
assemble[:half_num] = field[-half_num:]
assemble[half_num:] = field[:half_num]
```

```python
plt.plot(np.abs(assemble))
plt.show()
```

```python
plt.plot(assemble.real)
plt.show()
```

```python
plt.plot(assemble.imag)
plt.show()
```
