```python
import numpy as np
import matplotlib.pyplot as plt
import h5py as h5
import pickle
import scipy.ndimage as sn
```

```python
with open("C:/Users/hyli93/Desktop/Research/test/crystal_diff_test/axis_info_2019_07_01_19_09_11.pkl", 'rb') as axis_file:
    axis_info = pickle.load(axis_file)

with h5.File("C:/Users/hyli93/Desktop/Research/test/crystal_diff_test/result_2019_07_01_19_09_11.h5") as h5file:
    group = h5file['raw_data']
    
    labels = list(group.keys())
    for x in labels:
        print(x)
        
    data = np.array(group['electric field'])
```

```python
ratio = (axis_info['x_idx'][999] / 1000) / ( (axis_info['y_idx'][-1]  +  axis_info['y_idx'][-1])/ 500 )
```

```python
# Set up the canvas
fig, axes = plt.subplots(nrows=1, ncols=1)
fig.set_figheight(8)
fig.set_figwidth(20)

# Show the average pattern
im = axes.imshow( np.angle(np.fft.ifftshift(np.fft.ifft(data[:,:,0,2], axis=1), axes=1)),
                 #vmin= -5.,
                 #vmax = -5.,
                 cmap='jet', 
                 #origin='lower',
                 aspect = ratio,
                 )

#axes.set_title("Time response for R0H")
#axes.set_yticks(np.arange(0, 1000, step=200))
#axes.set_yticklabels(axis_info["x_idx_tick"][::200], fontsize=12)
#axes.set_xticks(np.arange(0, 500, step=50))
#axes.set_xticklabels(axis_info["y_idx_tick"][::50], fontsize=12)

axes.set_ylabel("Propagation direction (um)")
axes.set_xlabel("Transversal direction (um)")
fig.colorbar(im, ax=axes)
# Show the canvas

plt.show()
```
