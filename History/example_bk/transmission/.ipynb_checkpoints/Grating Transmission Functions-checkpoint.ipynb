# Define functions

```python
# This is for square grating. Used to check the new simulation
def transmission_m(k, m, n, h, a, b):
    """
    k: Wave number of the photon = 2 * pi * c / lambda
    m: The order of diffraction
    n: The complex refraction index.
    h: The height of the tooth.
    a: The width of the groove.
    b: The width of the tooth.
    """
    
    if not isinstance(m, int):
        raise Exception("m is the order of diffraction. This value has to be an integer.")
    if not isinstance(n, complex):
        raise Exception("n is the complex refraction index. This value has to be a complex number.")
    
    # First consider diffractions that are not the zeroth order.
    if m != 0:
        # Get the real part and the imaginary part of the refraction coefficient
        n_re = n.real - 1
        n_im = n.imag

        term_1 = 1 - np.exp(-k*h*n_im) * (np.cos(k*h*n_re) + 1.j * np.sin(k*h*n_re))
        term_2 = 1 - np.cos(2*np.pi*b*m/(a+b)) + np.sin(2*np.pi*b*m /(a+b))

        transmission = np.square(np.abs(term_1 * term_2)) / (4 * (np.pi * m)**2 )
        
    # Then consider the zeroth order
    else:
        # Get the real part and the imaginary part of the refraction coefficicent
        n_re = n.real - 1
        n_im = n.imag

        term_1 = (b + a * np.exp(-k*h*n_im) * (np.cos(k*h*n_re) + 1.j * np.sin(k*h*n_re))) / (a + b)

        transmission = np.square(np.abs(term_1))
    
    return transmission
```

```python
# This is for square grating. Used to check the new simulation
def transmission_triangle(k, m, n, h, a, b):
    """
    k: Wave number of the photon = 2 * pi * c / lambda
    m: The order of diffraction
    n: The complex refraction index.
    h: The height of the tooth.
    a: The width of the groove.
    b: The width of the tooth.
    """
    
    if not isinstance(m, int):
        raise Exception("m is the order of diffraction. This value has to be an integer.")
    if not isinstance(n, complex):
        raise Exception("n is the complex refraction index. This value has to be a complex number.")
    
    # First consider diffractions that are not the zeroth order.
    if m != 0:
        # Get the real part and the imaginary part of the refraction coefficient
        n_re = n.real - 1
        n_im = n.imag
        
        sh = k*h*(n - 1)
        
        exp_ish = np.exp(-k*h*n_im) * (np.cos(k*h*n_re) + 1.j * np.sin(k*h*n_re))

        exp_ia = np.cos(2 * np.pi * m * a / (a + b))  + 1.j * np.sin(2 * np.pi * a / (a + b))
        exp_ib = np.cos(2 * np.pi * m * b / (a + b))  + 1.j * np.sin(2 * np.pi * b / (a + b))
        
        transmission = 1. / (2 * np.pi * m + sh * (a + b) / a) * (exp_ish * exp_ia  - 1)
        transmission += 1. / (2 * np.pi * m - sh * (a + b) / b) * (1 - exp_ish * exp_ib)
        
        transmission = np.square(np.abs(transmission))
        
    # Then consider the zeroth order
    else:
        # Get the real part and the imaginary part of the refraction coefficicent
        n_re = n.real - 1
        n_im = n.imag

        sh = k*h*(n - 1)
        
        exp_ish = np.exp(-k*h*n_im) * (np.cos(k*h*n_re) + 1.j * np.sin(k*h*n_re))

        transmission = (exp_ish - 1) / sh
        
        transmission= np.square(np.abs(transmission))
    
    return transmission
```

```python
# Define function for the transmission efficiency
def transmission_0_1(k, m, n, h, a, b, c, d):
    """
    k: Wave number of the photon = 2 * pi * c / lambda
    m: The order of diffraction. This has to be 0 or 1
    n: The complex refraction index.
    h: The height of the tooth.
    a: The width of the groove.
    b: The width of the tooth.
    """
    
    if not ((m == 0) or (m ==1)):
        raise Exception("m has to be 1 or 0.")
        
    if not isinstance(n, complex):
        raise Exception("n is the complex refraction index. This value has to be a complex number.")
    
    
    # Get l
    l = a + b + c + d
    
    # First consider diffractions that are not the zeroth order.
    if m == 1:
        # Get the real part and the imaginary part of the refraction coefficient
        n_re = n.real - 1
        n_im = n.imag

        exp_ish = np.exp(-k*h*n_im) * (np.cos(k*h*n_re) + 1.j * np.sin(k*h*n_re))
        
        exp_2api = np.cos( 2 * np.pi * a / l ) + 1.j * np.sin( 2 * np.pi * a / l )
        exp_2bpi = np.cos( 2 * np.pi * b / l ) + 1.j * np.sin( 2 * np.pi * b / l )
        exp_2cpi = np.cos( 2 * np.pi * c / l ) + 1.j * np.sin( 2 * np.pi * c / l )
        exp_2dpi = np.cos( 2 * np.pi * d / l ) - 1.j * np.sin( 2 * np.pi * d / l )
        
        exp_2abpi = np.cos( 2 * np.pi * (a + b) / l ) + 1.j * np.sin( 2 * np.pi * (a + b) / l )
        
        first_term = ( exp_2api - 1 + exp_ish * exp_2abpi * (exp_2cpi - 1) ) * np.sqrt(l) / 2 / np.pi / 1.j
        
        second_term = (1 - exp_ish * exp_2dpi) * np.sqrt(l) * d / (2 * np.pi * d * 1.j - 1.j * (n-1) * k * h * l)
        
        thrid_term = (exp_ish * exp_2bpi - 1) * exp_2api  * np.sqrt(l) * b / (2 * np.pi * b * 1.j + 
                                                                              1.j * (n-1) * k * h * l)
        
        amplitude = first_term + second_term + thrid_term
        
    # Then consider the zeroth order
    if m == 0:
        # Get the real part and the imaginary part of the refraction coefficicent
        n_re = n.real - 1
        n_im = n.imag

        exp_ish = np.exp(-k*h*n_im) * (np.cos(k*h*n_re) + 1.j * np.sin(k*h*n_re))
        
        thrid_term = 1.j * (b + d) / ((n - 1) * k * h) * ( 1 -  exp_ish )
        
        amplitude = (a + c * exp_ish + thrid_term) / np.sqrt(l)
    
    return np.square(np.abs(amplitude) / np.sqrt(l))
```

# Check the simulation when a=c b=d=0

```python
# Speed of light * reduced Planck constant
c_h_bar = 2.99792458 * 6.582119569 * 1e-8

# Define photon energy and wave vector
photon_energy = 10000  #(eV)
k = photon_energy / c_h_bar

# Define the groove width
b = 1e-6  # (m)

# Define the tooth width
a = 1e-6  # (m)

# Define the refraction coefficent
n = 1 - 0.73031 * 1e-5 + 1.j * 0.61521 * 1e-8
#n = 1 -  7.29524527E-06 + 1.j * 7.61694174E-09

# Define a series of height
h = np.arange(1, 50) * 2e-7   # The unit is m. The smallest tooth height is 10nm.
```

```python
zero_trans_original = transmission_m(k = k, m = 0, n = n, h = h, a = a, b = b)
one_trans_original = transmission_m(k = k, m = 1, n = n, h = h, a = a, b = b)


zero_trans_new = transmission_0_1(k = k, m = 0, n = n, h = h, a = a, b = 0, c = b, d = 0)
one_trans_new = transmission_0_1(k = k, m = 1, n = n, h = h, a = a, b = 0, c = b, d = 0)
```

```python
plt.figure(figsize=(10, 10))

# Change the x axis unit into um
plt.plot(h * 1e6, zero_trans_new, 'r')
plt.plot(h * 1e6, one_trans_new, 'b')

plt.plot(h * 1e6, zero_trans_original, 'k*')
plt.plot(h * 1e6, one_trans_original, 'g*')


plt.xlabel("Hight of the tooth in um", fontsize=20)
plt.ylabel("Transmission efficiency", fontsize=20)
plt.legend(["0th order new", "1st order new", "0th order old", "1st order old"], fontsize=20)


plt.show()
```
